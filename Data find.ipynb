```python
import pandas as pd
```

```python
lookup = pd.read_csv("data/data.csv")
```

```python
lookup
```

```python
lookup = lookup.drop(lookup[lookup["length"] != 11].index)
```

```python
lookup
```

```python
temp = "etaoinsrhldcumfpgwybvkxjqz"
first_guess_list = []
for _ in temp:
    first_guess_list.append(_)
    
first_guess_list
```

```python
for letter in first_guess_list:
    if input(f"Does Your Word have {letter}") == "True":
        first_guess = letter
        break
```

```python
print(first_guess)
lookup.drop(lookup[lookup[f"{first_guess}"] == "-1"].index)
```

```python
number_of_first_guess = int(input(f"number of {first_guess}"))
location = []
for _ in range(number_of_first_guess):
    location.append(int(input(f"location of {first_guess}")))
location
```
